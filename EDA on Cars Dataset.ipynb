# Function for minimum and maximum observation

```python
def min_max_values(col):
    '''The function takes the ccolumn name as the argument
    and returns the top and bottom observation in that dataframe
    '''
    
    top = cars[col].idxmax()
    top_obs = pd.DataFrame(cars.loc[top])
    
    bottom = cars[col].idxmin()
    bottom_obs = pd.DataFrame(cars.loc[bottom])
    
    min_max_obs = pd.concat([top_obs,bottom_obs],axis=1)
    
    return min_max_obs
```

# Test the above function

```python
min_max_values('Fuel_efficiency')
```

# Plotting significant correlation in one plot

```python
corr = cars_num.drop('Fuel_efficiency',axis=1).corr()

sb.heatmap(corr[(corr >= 0.5) | (corr <= -0.4)],
          cmap='viridis', vmax=1.0,vmin=-1.0, linewidths=0.1,
          annot=True, annot_kws={'size':8},square=True)
```
